###Preenchimento de Arrays com números específicos

```python
#Array de uma dimensão
one_dimensional_array = np.array([1.2, 2.4, 3.5, 4.7, 6.1, 7.2, 8.3, 9.5])
one_dimensional_array
#print(one_dimensional_array)
```

```python
#Array de duas dimensões
two_dimensional_array = np.array([[6, 5], [11, 7], [4, 8]])
print(two_dimensional_array)
```

```python
# Preencher Array com zeros e uns
array_zeros = np.zeros(5)
print(array_zeros)
array_uns = np.ones([2,4])
print(array_uns)
```

```python
#Preencher Array com uma sequencia de numeros [lim_inferior, lim_superior)
#O intervalo é fechado no limite inferior e aberto no limite superior
sequence_of_integers = np.arange(5, 12)
print(sequence_of_integers)
```

```python
#preencher Array com números inteiros aleatórios
random_integers_between_50_and_100 = np.random.randint(low=50, high=101, size=(6))
print(random_integers_between_50_and_100)
```

```python
#preencher Array com números float aleatórios
random_floats_between_0_and_1 = np.random.random([6])
print(random_floats_between_0_and_1) 
```

###Operações Matemáticas com Numpy

```python
random_floats_between_2_and_3 = random_floats_between_0_and_1 + 2.0
print(random_floats_between_2_and_3)
```

```python
random_integers_between_150_and_300 = random_integers_between_50_and_100 * 3
print(random_integers_between_150_and_300)
```

#Tarefa: Criar um dataset linear


*   Criar um array "Feature" com uma sequencia de inteiros de 6 a 20 (inclusive)
*   Criar um array "Label" onde: 
   
   Label = (3)(feature) + 4


```python
#@title clique para ver a resposta

Feature = np.arange(6,21)
print(Feature)
Label = (Feature*3)+4
print(Label)
```



1.   Adicione algum ruído ao dataset

Para que o dataset pareça mais real, insira um ruído aleatório a cada elemento do array Label. Adicione um um valor ponto-flutuante aleatório entre -2 e 2.





```python
#@title Clique para ver a resposta

noise = (np.random.random([15]) * 4) - 2
print(noise)
Label = Label + noise 
print(Label)
```

```python
fig = plt.figure(figsize= (5, 4))

# Adicionando eixo a figura
ax = fig.add_axes([1, 1, 1, 1])
ax1 = ax.plot([1, 2, 3, 4], [1, 2, 3, 4])
ax2 = ax.plot([1, 2, 3, 4], [2, 3, 4, 5])
```

```python
x = [3, 1, 3]
y = [3, 2, 1]

fig = plt.figure(figsize =(5, 4))
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x, y)
ax.set_xlim(1, 2)
# mudando labes do eixo X
ax.set_xticklabels(("One", "Two", "Three", "Four", "Five", "Six"))
```

```python
# Adicionando titulo e label(legenda) ao grafico
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], label = "dados")
plt.axis([0, 6, 0, 20])
plt.title('Meu gráfico')
plt.legend()
plt.show()
```

```python
# Criando figura com 4 graficos e gerando graficos aleatorios, definindo cor da linha, titulo da figura e do grafico e legenda em cada grafico
fig1, f1_axes = plt.subplots(ncols = 2, nrows = 2, figsize = (10, 10))
fig1.suptitle("Meus gráficos", size = 20)

f1_axes[0,0].plot(np.random.random(5), np.random.random(5), label = "Valores", color = "red")
f1_axes[0,0].set_title("Gráfico 1")
f1_axes[0,0].set_xlabel("Label eixo X")
f1_axes[0,0].set_ylabel("Label eixo Y")
f1_axes[0,0].legend()

f1_axes[0,1].plot(np.random.random(5), np.random.random(5), label = "Valores", color = "blue")
f1_axes[0,1].set_title("Gráfico 2")
f1_axes[0,1].set_xlabel("Label eixo X")
f1_axes[0,1].set_ylabel("Label eixo Y")
f1_axes[0,1].legend()

f1_axes[1,0].plot(np.random.random(5), np.random.random(5), label = "Valores", color = "green")
f1_axes[1,0].set_title("Gráfico 3")
f1_axes[1,0].set_xlabel("Label eixo X")
f1_axes[1,0].set_ylabel("Label eixo Y")
f1_axes[1,0].legend()

f1_axes[1,1].plot(np.random.random(5), np.random.random(5), label = "Valores", color = "yellow")
f1_axes[1,1].set_title("Gráfico 4")
f1_axes[1,1].set_xlabel("Label eixo X")
f1_axes[1,1].set_ylabel("Label eixo Y")
f1_axes[1,1].legend()
```
