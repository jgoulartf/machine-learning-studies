**loc**

Pode-se usar o método loc para localizar linhas ou pedaços do dataframe pelos rótulos

Sintaxe:

loc[linhas, colunas]

Onde as linhas e colunas podem ser passadas em sequencia (:) ou saltos (,)

```python
df.loc[0, 'cargo'] #um elemento
```

```python
df.loc[0:3] #uma sequencia de linhas
```

```python
df.loc[0:3, 'cargo':'salario'] #pedaço do dataframe, usando sequencia de linhas e colunas
```

```python
df.loc[[0,3], ['cargo', 'salario']] #pedaço do dataframe, usando linhas e colunas em saltos (observar colchetes)
```

```python
#como pegar duas sequências diferentes:

pd.concat([df.loc[0:3], df.loc[10:12]]) 

#pode usar o concat do pandas que irá concatenar dois dataframes (ou pedaços de dataframes)
```

**iloc**

Pode-se usar o método iloc para localizar linhas ou pedaços do dataframe pelos índices numéricos (posicionais)

Sintaxe (igual a do loc):

iloc[linhas, colunas]

Onde as linhas e colunas podem usar sequencias (:) ou saltos (,)

A diferença de iloc para loc é que loc usa os rotulos, enquanto iloc usa os indices numericos default

```python
df.iloc[0, 0] #um elemento (linha 0, coluna cargo)
```

```python
df.iloc[0:3, 0:3] #sequencia
```

```python
df.iloc[[0,3], [0,3]] #salto
```

###Filtros no Data Frame

#### 1 - Diretamente nas colunas

```python
#Filtrando valores em uma coluna especifica
df['idade']>25
```

```python
# guardando o resultado do filtro em outra variavel
df_idade = df[df['idade']>25]
print(df_idade)
# apresentando o formato da variavel
df_idade.shape
```

```python
df[(df['idade']>25) & (df['salario']>7000)]
```

#### 2 - Usando o método query

```python
#Outra forma de filtrar dados no Dataset é por meio do método query
df.query('idade >25')
```

```python
df.query('idade >25 & salario>7000')
```

```python
# usando uma variavel como parametro na query
idade_mediana = df['idade'].median()
print(idade_mediana)
df.query('idade > @idade_mediana')
```

```python
#query com string
df.query('cargo == "Diretor"')
```

```python
# o metodo isin seleciona os registros que atendem aos itens listados
df[df['cargo'].isin(['Analista','Programador'])]
```

#### 3 - Utilizando o método filter

```python
df.filter(['cargo','idade','salario'])
```

### Métodos úteis
Value_counts

retorna uma série contendo contagens de valores únicos. O objeto resultante estará em ordem decrescente para que o primeiro elemento seja o elemento que ocorre com mais frequência. Exclui valores NA por padrão.


```python
#apresenta a quantidade de valores unicos para uma coluna em particular
df['cargo'].value_counts()
```

```python
#Para atributos categoricos podemos usar o parametro bins
df['salario'].value_counts(bins=4)
```

```python
#apresenta a quantidade de valores em cada coluna
df.count()
```

####AGREGANDO INFORMAÇÕES NO GROUPBY
